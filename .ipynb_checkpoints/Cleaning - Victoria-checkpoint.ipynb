```python
import pandas as pd
df = pd.read_csv("comments2.csv") 
```

```python
df = df.drop("Unnamed: 0", axis=1)
df = df.drop("link", axis=1)
df = df.dropna()
```

```python
df = df[df.body != '[deleted]']
df = df[df.body != '[removed]']
```

```python
df = df.replace(r"(?:\@|https?\://)\S+", "", regex=True)
df = df[df.body != '']
```
